```python
var1 = 'comb'
var2 = 'co2'
# set bin edges, compute centers
bin_size = 0.25
xbin_edges = np.arange(0.5, df[var1].max()+bin_size, bin_size)
xbin_centers = (xbin_edges + bin_size/2)[:-1]

# compute statistics in each bin
data_xbins = pd.cut(df[var1], xbin_edges, right = False, include_lowest = True)
y_means = df[var2].groupby(data_xbins).mean()
y_sems = df[var2].groupby(data_xbins).sem()

# plot the summarized data
plt.errorbar(x = xbin_centers, y = y_means, yerr = y_sems)
plt.xlabel(var1)
plt.ylabel(var2)
```
