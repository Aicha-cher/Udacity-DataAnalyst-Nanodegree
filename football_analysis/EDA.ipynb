```python
list_of_team_id = df_team_to_player_2015.home_team_api_id.unique().tolist()
```

```python
team_dict = {}
```

```python
for team_name in list_of_team_id:
    df = df_team_to_player_2015[df_team_to_player_2015.home_team_api_id== team_name]
    for player_name in pd.Series(df.values.ravel('F')).unique():
        team_dict[player_name]=team_name                           
```

```python
df_player_team_2015 = pd.DataFrame(team_dict,index=[0])
df_player_team_2015=df_player_team_2015.T
df_player_team_2015.reset_index(level=0, inplace=True)
df_player_team_2015.rename(index=str, columns={"index": "player_api_id", 0: "team_api_id"},inplace=True)
df_player_team_2015.head()
```
