```python
"""
@author: abhik
"""

#imports 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_global = pd.read_csv('results_global.csv')
df_hyd = pd.read_csv('results_hyd_IN.csv')
```

```python
df_global.isna().sum() #counting missing data 
df_global.head()
```

```python
df_global.isna().sum() #counting missing data 
df_hyd.head()
```

```python
df_hyd = df_hyd.interpolate(method='spline', order=2)
```

```python
df_hyd.plot(x='year',y='avg_temp',kind='line') #kind='box' kind='hist' 
df_global.plot(x='year',y='avg_temp',kind='line')
plt.show()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula="avg_temp ~ year", data=df_hyd).fit()
trend = result.predict(df_hyd.year)
data = pd.DataFrame(index=df_hyd.year.values, data={'y': df_hyd.avg_temp.values, 'trend': trend.values})
data.plot() 
plt.show()
```

```python
result = sm.ols(formula="avg_temp ~ year", data=df_global).fit()
trend = result.predict(df_global.year)
data = pd.DataFrame(index=df_global.year.values, data={'y': df_global.avg_temp.values, 'trend': trend.values})
data.plot() 
plt.show()
```

```python
df_global = df_global.set_index('year')
df_hyd = df_hyd.set_index('year')
```

```python
df_plot = pd.merge(df_hyd, df_global, on="year")
del df_plot['country']
del df_plot['city']
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df_plot)
```
